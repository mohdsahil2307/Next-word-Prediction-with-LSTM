```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
with open('hamlet.txt','r') as file:
  text = file.read().lower()
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text]) #Creates a word index mapping
```

```python
total_words = len(tokenizer.word_index) +1
total_words
```

```python
input_sequences = []
count = 0
for line in text.split('\n'):
  token_list = tokenizer.texts_to_sequences([line])[0]
  for i in range(len(token_list)):
    n_grams = token_list[:i+1]
    input_sequences.append(n_grams)
```

```python
input_sequences
```

```python
max_sequence_len = max([len(x) for x in input_sequences])
max_sequence_len
```

```python
import numpy as np
input_sequences = np.array(pad_sequences(input_sequences,maxlen=max_sequence_len,padding='pre'))
```

```python
X,y = input_sequences[:,:-1],input_sequences[:,-1]
```

```python
import tensorflow as tf
y = tf.keras.utils.to_categorical(y,num_classes=total_words)
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout
model = Sequential()
model.add(Embedding(total_words,100,input_length = max_sequence_len-1))
model.add(LSTM(150,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words,activation='softmax'))

model.compile(loss='categorical_crossentropy',metrics= ['accuracy'], optimizer='adam')
model.summary()
```

```python
## MODEL TRAININGGGGG
history = model.fit(X_train,y_train,epochs=50,validation_data=(X_test,y_test),verbose=1)
```

```python
model.summary()
```

```python
def predict_with_text(model,text):
  tokenized_text = tokenizer.texts_to_sequences([text])[0]
  if len(tokenized_text) > max_sequence_len:
    tokenized_text = tokenized_text[-(max_sequence_len-1):]
  tokenized_text = pad_sequences([tokenized_text],maxlen=max_sequence_len-1,padding='pre')
  prediction = model.predict(tokenized_text,verbose=0)
  most_probable = np.argmax(prediction,axis=1)
  for word,index in tokenizer.word_index.items():
    if index == most_probable:
      return word
  return None

print(predict_with_text(model,'To be or not to be'))
```
